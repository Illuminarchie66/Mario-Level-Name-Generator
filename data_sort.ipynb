```python
import json 
from pprint import pprint
```

```python
with open('named_levels.json') as f:
    all_levels = json.load(f)
```

```python
nsmbu = [level for level in all_levels if level.get("meta", {}).get("game") == "New Super Mario Bros. U"]
nslbu = [level for level in all_levels if level.get("meta", {}).get("game") == "New Super Luigi U"]
smb3dw = [level for level in all_levels if level.get("meta", {}).get("game") == "Super Mario 3D World"]
cttt = [level for level in all_levels if level.get("meta", {}).get("game") == "Captain Toad's Treasure Tracker"]

nsmbds = [level for level in all_levels if level.get("meta", {}).get("game") == "New Super Mario Bros."]
nsmbwii = [level for level in all_levels if level.get("meta", {}).get("game") == "New Super Mario Bros. Wii"]
nsmb2 = [level for level in all_levels if level.get("meta", {}).get("game") == "New Super Mario Bros. 2"]

not_in_any = [level for level in all_levels if level.get("meta", {}).get("game") not in ["New Super Mario Bros. U", "New Super Luigi U", "Super Mario 3D World", "Captain Toad's Treasure Tracker", "New Super Mario Bros.", "New Super Mario Bros. Wii", "New Super Mario Bros. 2"]]
```

```python
print(not_in_any)
```

```python
print(f"""
Total levels: {len(all_levels)}

--- Named levels ---
NSMBU: {len(nsmbu)}
NSLBU: {len(nslbu)}
SMB3DW: {len(smb3dw)}
CTTT: {len(cttt)}

--- Unnamed levels ---
NSMBDS: {len(nsmbds)}
NSMBWII: {len(nsmbwii)}
NSMB2: {len(nsmb2)}

--- Misc ---
Not in any: {len(not_in_any)}
      
""")
```

```python
with open("data/named/nsmbu.json", "w") as f:
    json.dump(nsmbu, f, indent=4)
with open("data/named/nslbu.json", "w") as f:
    json.dump(nslbu, f, indent=4)
with open("data/named/smb3dw.json", "w") as f:
    json.dump(smb3dw, f, indent=4)
with open("data/named/cttt.json", "w") as f:
    json.dump(cttt, f, indent=4)

with open("data/unnamed/nsmbds.json", "w") as f:
    json.dump(nsmbds, f, indent=4)
with open("data/unnamed/nsmbwii.json", "w") as f:
    json.dump(nsmbwii, f, indent=4)
with open("data/unnamed/nsmb2.json", "w") as f:
    json.dump(nsmb2, f, indent=4)
```
