```python
import json 
from pprint import pprint
```

```python
with open("data/named/nsmbu.json", "r") as f:
    nsmbu = json.load(f)
with open("data/named/nslbu.json", "r") as f:
    nslbu = json.load(f)
with open("data/named/smb3dw.json", "r") as f:
    smb3dw = json.load(f)
with open("data/named/cttt.json", "r") as f:
    cttt = json.load(f)

with open("data/unnamed/nsmbds.json", "r") as f:
    nsmbds = json.load(f)
with open("data/unnamed/nsmbwii.json", "r") as f:
    nsmbwii = json.load(f)
with open("data/unnamed/nsmb2.json", "r") as f:
    nsmb2 = json.load(f)
```

```python
print(f"""
Total levels: {len(nsmbu) + len(nslbu) + len(smb3dw) + len(cttt) + len(nsmbds) + len(nsmbwii) + len(nsmb2)}

--- Named levels ---
NSMBU: {len(nsmbu)}
NSLBU: {len(nslbu)}
SMB3DW: {len(smb3dw)}
CTTT: {len(cttt)}

--- Unnamed levels ---
NSMBDS: {len(nsmbds)}
NSMBWII: {len(nsmbwii)}
NSMB2: {len(nsmb2)}
      
""")
```

```python
with open("data/named/nsmbu.json", "w") as f:
    json.dump(nsmbu, f, indent=4)
with open("data/named/nslbu.json", "w") as f:
    json.dump(nslbu, f, indent=4)
with open("data/named/smb3dw.json", "w") as f:
    json.dump(smb3dw, f, indent=4)
with open("data/named/cttt.json", "w") as f:
    json.dump(cttt, f, indent=4)

with open("data/unnamed/nsmbds.json", "w") as f:
    json.dump(nsmbds, f, indent=4)
with open("data/unnamed/nsmbwii.json", "w") as f:
    json.dump(nsmbwii, f, indent=4)
with open("data/unnamed/nsmb2.json", "w") as f:
    json.dump(nsmb2, f, indent=4)
```
